```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('iris.csv')
```

```python
dataset
```

```python
X=dataset.iloc[:,1:-1].values
X
```

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(X,method="ward"))
plt.title("Dendrogram")
plt.xlabel("Sepal")
plt.ylabel("Eucllidean distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
clustering=AgglomerativeClustering(n_clusters=3)
y_hc=clustering.fit_predict(X)
```

```python
y_hc
```

```python
plt.scatter(X[y_hc==0,0],X[y_hc==0,2],c='red',label='Cluster1')
plt.scatter(X[y_hc==0,1],X[y_hc==0,2],c='blue',label='Cluster2')
plt.scatter(X[y_hc==0,2],X[y_hc==0,2],c='green',label='Cluster3')
plt.title("Clusters of  iris flower")
plt.ylabel("Petal Length(cm)")
plt.xlabel("Sepal Length(cm)")
plt.legend()
plt.show()
```
